```python
import datetime
def datelist(start, end):
    start_date = datetime.date(*start)
    end_date = datetime.date(*end)

    result = []
    curr_date = start_date
    while curr_date != end_date:
        result.append("%04d%02d%02d" % (curr_date.year, curr_date.month, curr_date.day))
        curr_date += datetime.timedelta(1)
    result.append("%04d%02d%02d" % (curr_date.year, curr_date.month, curr_date.day))
    return result
date = datelist((2020, 1, 1), (2022, 1, 1))
#go through press release between 2020-1-1 and 2022-1-1
```

```python
from bs4 import BeautifulSoup
import urllib.request
#from urllib.request import Request

seed_url = "https://www.federalreserve.gov/newsevents/pressreleases/"
child_url = []
for i in date:
    child = seed_url+'bcreg'+ str(i)+ 'a.htm'
    #print(child_url)
    child_url.append(child)
```

```python
seen = []    #stack of urls seen so far
opened = []          #we keep track of seen urls so that we don't revisit them
final_url = []

maxNumUrl = 1000; #set the maximum number of urls to visit
while len(child_url) > 0 and len(opened) < maxNumUrl:
    # DEQUEUE A URL FROM urls AND TRY TO OPEN AND READ IT
    try:
        curr_url=child_url.pop(0)
        #print("num. of URLs in stack: %d " % len(urls))
        print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(curr_url)

    except Exception as ex:
        #print("Unable to access= "+curr_url)
        #print(ex)
        continue    #skip code below

    # IF URL OPENS, CHECK WHICH URLS THE PAGE CONTAINS
    # ADD THE URLS FOUND TO THE QUEUE url AND seen
    soup = BeautifulSoup(webpage)  #creates object soup
    # Put child URLs into the stack
    for tag in soup.find_all('div', {'class':'col-xs-12 col-sm-8 col-md-8'}): #find tags with links
        text = tag.get_text().lower()
        #print(text)
        if 'covid' in text:
            final_url.append(curr_url)
        else:
            continue
    if len(final_url)>10:
        break
```
